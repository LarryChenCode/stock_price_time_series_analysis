# **2. Download the adjusted close prices for FB, MMM, IBM and AMZN for the last 60 months**

# **3. Resample the data to get prices for the end of the business month. Select the Adjusted Close for each stock.**

```python
#Resmaple META to get adjusted close prices for the end of the business month
meta_resampled = meta['Adj Close'].resample("BM").last().to_frame()
meta_resampled
```

```python
#Resmaple MMM to get adjusted close prices for the end of the business month
mmm_resampled = mmm['Adj Close'].resample("BM").last().to_frame()
mmm_resampled
```

```python
#Resmaple IBM to get adjusted close prices for the end of the business month
ibm_resampled = ibm['Adj Close'].resample("BM").last().to_frame()
ibm_resampled
```

```python
#Resmaple AMZN to get adjusted close prices for the end of the business month
amzn_resampled = amzn['Adj Close'].resample("BM").last().to_frame()
amzn_resampled
```

# **4. Use the pandas autocorrelation_plot() function to plot the autocorrelation of the adjusted month-end close prices for each of the stocks.**









```python
#plot the autocorrelation of the adjusted month-end close prices for META
pd.plotting.autocorrelation_plot(series=meta_resampled)
```

```python
#plot the autocorrelation of the adjusted month-end close prices for MMM
pd.plotting.autocorrelation_plot(series=mmm_resampled)
```

```python
#plot the autocorrelation of the adjusted month-end close prices for IBM
pd.plotting.autocorrelation_plot(series=ibm_resampled)
```

```python
#plot the autocorrelation of the adjusted month-end close prices for AMZN
pd.plotting.autocorrelation_plot(series=amzn_resampled)
```

Upon analyzing the plots of the META, MMM, IBM, and META stocks shown below, it becomes evident that the adjusted month-end closing prices exhibit highly positive autocorrelation initially. However, over time, the correlation gradually starts oscillating between positive and negative values, indicating a diminishing trend in correlation. Eventually, the autocorrelation reaches 0 when the lag is 60.

# **5.**

- **Calculate the monthly returns for each stock using the "shift trick" explained in the lecture, using shift() function.**

- **Use pandas autotocorrelation_plot() to plot the autocorrelation of the monthly returns.**
  


```python
#META
meta_resampled['lag1'] = meta_resampled['Adj Close'].shift(1)
meta_resampled['Monthly Returns'] = meta_resampled['Adj Close'] - meta_resampled['lag1']
pd.plotting.autocorrelation_plot(series=meta_resampled['Monthly Returns'].dropna())
```

```python
#MMM
mmm_resampled['lag1'] = mmm_resampled['Adj Close'].shift(1)
mmm_resampled['Monthly Returns'] = mmm_resampled['Adj Close'] - mmm_resampled['lag1']
pd.plotting.autocorrelation_plot(series=mmm_resampled['Monthly Returns'].dropna())
```

```python
#IBM
ibm_resampled['lag1'] = ibm_resampled['Adj Close'].shift(1)
ibm_resampled['Monthly Returns'] = ibm_resampled['Adj Close'] - ibm_resampled['lag1']
pd.plotting.autocorrelation_plot(series=ibm_resampled['Monthly Returns'].dropna())
```

```python
#AMZN
amzn_resampled['lag1'] = amzn_resampled['Adj Close'].shift(1)
amzn_resampled['Monthly Returns'] = amzn_resampled['Adj Close'] - amzn_resampled['lag1']
pd.plotting.autocorrelation_plot(series=amzn_resampled['Monthly Returns'].dropna())
```

# **6.**

- **Combine all 4 time series (returns) into a single DataFrame,**

- **Visualize the correlation between the returns of all pairs of stocks using a scatter plot matrix (use scatter_matrix() function from pandas.plotting).**

```python
#Combine all 4 time series (returns) into a single DataFrame
four_stks = pd.concat([meta_resampled['Monthly Returns'], mmm_resampled['Monthly Returns'], ibm_resampled['Monthly Returns'], amzn_resampled['Monthly Returns']], axis=1).dropna()
```

```python
four_stks_returns = four_stks.set_axis(['META', 'MMM', 'IBM', 'AMZN'], axis=1, inplace=False)
four_stks_returns
```

```python
#Visualize the correlation - scatter_matrix() function
pd.plotting.scatter_matrix(four_stks_returns)
```

```python
#Show correlation
four_stks_returns.corr()
```

  
  Upon applying the scatter_matrix() and corr() functions, it becomes evident that all combinations are positively correlated, but not significantly strong. Only the IBM vs MMM combination has a correlation above 0.50, while the values for the other combinations range from 0.45 to 0.17.
